```python
# Contrastive loss
def contrastive_loss(y_true, y_pred):
    """Contrastive loss function.

    Args:
        y_true: Ground truth labels.
        y_pred: Predicted labels.

    Returns:
        Loss value.
    """
    y_true = tf.cast(y_true, tf.float32)  # Cast y_true to float32
    loss = tf.reduce_mean(tf.square(y_true - y_pred))
    return loss
```

# methods

```python
tokenizer = AutoTokenizer.from_pretrained('bert-large-uncased') # BertTokenizer.from_pretrained
bertmodel = BertModel.from_pretrained('bert-large-uncased')
```

```python
# Contrastive loss
def contrastive_loss(y_true, y_pred):
    """Contrastive loss function.

    Args:
        y_true: Ground truth labels.
        y_pred: Predicted labels.

    Returns:
        Loss value.
    """
    y_true = tf.cast(y_true, tf.float32)  # Cast y_true to float32
    loss = tf.reduce_mean(tf.square(y_true - y_pred))
    return loss
```

```python
def preprocess(text):
    nlp = spacy.load("en_core_web_sm")
    return nlp(text).text
```

```python
def generateEmbeddingpair(context, response):

    input_ids_1 = tokenizer.encode(context, return_tensors='pt',max_length=512, truncation=True)
    last_hidden_states_1 = bertmodel(input_ids_1)[0]
    context_embedding = last_hidden_states_1[0][0].detach().numpy()


    input_ids_2 = tokenizer.encode(response, return_tensors='pt',max_length=512, truncation=True)
    last_hidden_states_2 = bertmodel(input_ids_2)[0]
    response_embedding = last_hidden_states_2[0][0].detach().numpy()

    embeddings_pair = np.array(list(zip(context_embedding, response_embedding)))
    # print(embeddings_pair.shape)
    embeddings_pair = embeddings_pair.reshape(1, embeddings_pair.shape[1],embeddings_pair.shape[0])
    # print(embeddings_pair.shape)
    return embeddings_pair
```

```python
def get_predicted_labels(output, threshold):
    # print(output, threshold)
    if(output < threshold):
        return 'NOT SARCASM'
    return 'SARCASM'
    # x = (output >= threshold).astype(int)
    # ans = 'SARCASM' if x==1 else
    # return ans
```

# All models

```python
import os
def sarcasm_detection_all(context, response):
    listy = []
    embeddings_pair = generateEmbeddingpair(preprocess(context), preprocess(response))
    path = '/content/gdrive/MyDrive/embeddings/MODELS/'
    for modelname in os.listdir(path):
        if "model" not in modelname:
            continue
        model = tf.keras.models.load_model(path+modelname, custom_objects={'contrastive_loss': contrastive_loss})

        ans = model.predict(x=[embeddings_pair[:,0], embeddings_pair[:,1]])

        threshold = 0.25  # Use the same F1-score threshold as during training
        predicted_labels = get_predicted_labels(ans[0][0], threshold)
        listy.append(modelname+ ':' + str(ans) + '==> ' + predicted_labels)

        # print(predicted_labels, '\n')
        # return predicted_labels
    return listy
```
